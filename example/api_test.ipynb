```python
import pandas as pd
import requests
```

```python

# Define the API endpoint and parameters
url = "http://services.cehub.syngenta-ais.com/api/Forecast/ShortRangeForecastHourly"
params = {
    "wkt": "point(9.37477 47.42391)", # lng lat
    "startDate": "2025-03-20",
    "endDate": "2025-03-21",
    "supplier": "Meteoblue",
    "measureLabel": "SunshineDuration_Hourly (min)",
    "top": 500,
    "format": "json",
    "ApiKey": "d4f087c7-7efc-41b4-9292-0f22b6199215"  # Replace with your actual API key
}

# Make the request
response = requests.get(url, params=params)

# Check the response status
if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

```python
import json

# Define the API endpoint and parameters
url = "http://my.meteoblue.com/dataset/query?apikey=7b29a207a0de"
headers = {
    "Content-Type": "application/json"
}
data = {
    "units": {
        "temperature": "C",
        "velocity": "km/h",
        "length": "metric",
        "energy": "watts"
    },
    "geometry": {
        "type": "MultiPoint",
        "coordinates": [[7.57327, 47.558399, 279]],  # lon, lat, asl
        "locationNames": ["Basel"]
    },
    "format": "json",
    "timeIntervals": [
        "2019-01-01T+00:00/2019-12-31T+00:00"
    ],
    "queries": [{
        "domain": "NEMSGLOBAL",
        "gapFillDomain": None,
        "timeResolution": "hourly",
        "codes": [{
            "code": 157,
            "level": "180-0 mb above gnd"
        }]
    }]
}

# Make the request
response = requests.post(url, headers=headers, data=json.dumps(data))

# Check the response status
if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```
