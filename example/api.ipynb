```python
import requests
import pandas as pd

def get_weather_forecast(forecast_type, measure_labels, start_date, end_date, api_key):
    # Define the API endpoint
    url = "http://services.cehub.syngenta-ais.com/api/Forecast/ShortRangeForecastHourly"

    # Define the parameters
    params = {
        "wkt": "point(9.37477 47.42391)",  # point(lng lat)
        "startDate": start_date,  # YYYY-MM-DD
        "endDate": end_date,  # YYYY-MM-DD
        "supplier": "Meteoblue",
        "measureLabel": ";".join(measure_labels),
        "top": 500,
        "format": "json",
        "ApiKey": api_key  # Replace with your actual API key
    }

    # Make the request
    response = requests.get(url, params=params)

    # Check the response status
    if response.status_code == 200:
        data = pd.DataFrame(response.json())
        return data
    else:
        print(f"Error: {response.status_code}")
        print(response.text)
        return None

# Example usage
forecast_type = "Hourly"  # Choose between "Nowcast", "Hourly", "Daily"
measure_labels = [
    "Cloudcover_Hourly (pct)",
    "GlobalRadiation_HourlySum (Wh/m2)",
    "HumidityRel_Hourly (pct)"
    # Add more labels as needed
]
start_date = "2025-03-20"
end_date = "2025-03-21"
api_key = "api_key_here"  # Replace with your actual API key

data = get_weather_forecast(forecast_type, measure_labels, start_date, end_date, api_key)
if data is not None:
    display(data)
```
